```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,label="Sine Wave")
plt.xlabel("X Axis")
plt.xlabel("Y Axis")
plt.ylabel("simple Line Plot")
plt.legend()
plt.show()
```
